```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Assuming the data is in a CSV file named "Data_Salaries.csv"
df = pd.read_csv("Data_Salaries.csv")
```

```python
# Data Cleaning
# Drop any rows with missing values
df.dropna(inplace=True)
```

```python
# Descriptive Statistics
print(df.describe())
```

```python
from wordcloud import WordCloud

# Filtering top 20 companies with a score greater than 4.5
top_companies = df[df['Company Score'] > 4.5][['Company', 'Company Score']].sort_values(by='Company Score', ascending=False).head(20)

# Create a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(top_companies['Company']))

# Visualization: Word Cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Top 20 Companies with a Score Greater than 4.5')
plt.show()
```

```python
# Visualization 2: Company Score vs. Salary
plt.figure(figsize=(10, 90))
sns.scatterplot(x="Company Score", y="Salary", data=df)
plt.xlabel("Company Score")
plt.ylabel("Salary")
plt.title("Company Score vs. Salary")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Company Score'])
plt.title('Distribution of Company Scores')
plt.xlabel('Company Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# The average salary per company
avg_salary_by_company = df.groupby('Company')['Monthly'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 8))
sns.barplot(x=avg_salary_by_company.values, y=avg_salary_by_company.index, palette='nipy_spectral_r')
plt.title('Top 10 Companies with Highest Average Salary', fontsize=16)
plt.xlabel('Average Salary monthly', fontsize=14)
plt.ylabel('Company', fontsize=14)
plt.show()
```
